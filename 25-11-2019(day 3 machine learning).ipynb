#### CRIM: This is the per capita crime rate by town
#### ZN: This is the proportion of residential land zoned for lots larger than 25,000 sq.ft.

#### CHAS: This is the Charles River dummy variable (this is equal to 1 if tract bounds river; 0 otherwise)
#### NOX: This is the nitric oxides concentration (parts per 10 million)
#### RM: This is the average number of rooms per dwelling
#### AGE: This is the proportion of owner-occupied units built prior to 1940
#### DIS: This is the weighted distances to five Boston employment centers
#### RAD: This is the index of accessibility to radial highways
#### TAX: This is the full-value property-tax rate per 10,000
#### PTRATIO: This is the pupil-teacher ratio by town
#### B: This is calculated as 1000(Bk — 0.63)², where Bk is the proportion of people of African American descent by town
#### LSTAT: This is the percentage lower status of the population
#### MEDV: This is the median value of owner-occupied homes in 1000s

```python
# prepare the data for training

X = pd.DataFrame(np.c_[boston['LSTAT'],boston['RM']], columns = ['LSTAT',"RM"])
print(X)
Y = boston['MEDV']
print(Y)
```

```python
from sklearn.model_selection import train_test_split
a, b = np.arange(10).reshape((5,2)),range(5)
#print(a)
#print(b)
train_test_split(a,b,random_state = 1)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.2,random_state=5)
print(X_train.shape)
print(X_test.shape)
print(X_test)
print(Y_train.shape)
print(Y_test.shape)
print(Y_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
```

```python
# model evaluation for traing set

y_train_predict = lin_model.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train,y_train_predict)))
r2 = r2_score(Y_train, y_train_predict)
print("the model performance for training set")
print("--------------------------------------")
print("RMSE is {}".format(rmse))
print("R2 score is {}".format(r2))

# model evaluation for testing set

y_test_predict = lin_model.predict(X_test)
# root mean squre error of the model
rmse = (np.sqrt(mean_squared_error(Y_test,y_test_predict)))

# r-squred score of the model
r2 = r2_score(Y_test, y_test_predict)

print("the model performance for testing set")
print("-------------------------------------")
print("RMSE is {}".format(rmse))
print("R2 score is {}".format(r2))
```

```python
# platting the y_test and y_pred
# ideally should have been a straight line
plt.scatter(Y_test, y_test_predict)
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree = 2)
# transforms the features to higher degree functions
X_train_poly = poly_features.fit_transform(X_train)
# fit the transformed fatures to linear regression
poly_model = LinearRegression()
poly_model.fit(X_train_poly,Y_train)

# prediction on traing data-set
y_train_predict = poly_model.predict(X_train_poly)

# predection on test data sets
y_test_predict = poly_model.predict(poly_features.fit_transform(X_test))

# evaluating the model on training dataset
rmse_train = np.sqrt(mean_squared_error(Y_train, y_train_predict))
r2_train = r2_score(Y_train, y_train_predict)

# evaluating the model on test dataset
rmse_test = np.sqrt(mean_squared_error(Y_test,y_test_predict))
r2_test = r2_score(Y_test, y_test_predict)
print("Model performnace for the training set")
print("--------------------------------------")
print("RMSE of training set  is {}".format(rmse_train))
print("R2 score of tarining set is {}".format(r2_train))
print("\n")

print("Model performnace forr the testing set")
print("--------------------------------------")
print("RMSE of testing set is  {}".format(rmse_test))
print("R2 score of testing set is {}".format(r2_test))
```

## Ridge Rgression

```python
from sklearn.linear_model import Ridge
```

```python
ridge_reg = Ridge(alpha = 1)#, solver = 'cholesky'
ridge_reg
# the L2 norm term in ridge regression is weighted by the regularization parameter alpha
# so if the alpha values is 0, it means that it is just an ordinary least squares regression model
# so the larger is the alpha the higher is the smoothness contraints
```

```python
import numpy as np
m =100
X = 6 * np.random.randn(m,1)-3
y = 0.5 * X**2 + X + 2 + np.random.randn(m,1)
ridge_reg.fit(X,y)
```

```python
print(ridge_reg.intercept_,ridge_reg.coef_)
print(ridge_reg.predict([[1.5]]))
```

```python
from sklearn.linear_model import Lasso
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
print(cancer.data.shape)
```

```python
cancer_df = pd.DataFrame(cancer.data,columns = cancer.feature_names)
```

```python
cancer_df
```

```python
X = cancer.data
Y = cancer.target
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.1,random_state=31)
```

```python
lasso = Lasso()
lasso.fit(X_train,Y_train)
```

```python
train_score = lasso.score(X_train,Y_train)
test_score = lasso.score(X_test,Y_test)
print(train_score)
print(test_score)
```

```python
coeff_used = np.sum(lasso.coef_!=0)
coeff_used
```

```python
print("Training score",train_score)
print("Testing score",test_score)
print("Number of features used:",coeff_used)
```
